```python
# importar librerias
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Para este ejercicio, primero debe obtener los datos de gapminder de la URL 
# de carpintería de software y cargarlos como marco de datos de pandas. Los 
# datos de gapminder tienen información sobre la vida útil, la población y el pib 
# de los países durante varios años.

gapminder_csv_url ='http://bit.ly/2cLzoxH'
gapminder = pd.read_csv(gapminder_csv_url)
gapminder
```

```python
# 1. Realice un resumen estadístico de sus datos.

gapminder.describe() 
```

```python
# 2. Agrupe por la variable “año”. Comente respecto al resultado al usar groupby.

gapminder.groupby('year')

# COMENTARIO: Al agrupar por año, se obtiene un objeto de tipo DataFrameGroupBy, 
# se están creando grupos de datos separados para cada año.   
```

```python
# 3. Ahora obtenga la suma (la cantidad), la media y la mediana para cada grupo, es decir, por año.
gapminder.groupby('year').aggregate(['sum','mean','median'])
```

```python
# 4. ¿Sabías que en 1967 en Chile se inaugura el Aeropuerto Internacional 
# Arturo Merino Benítez? Usando estos datos se puede especificar un año 
# y obtener un marco de datos más pequeño para el año usando la función 
# get_group(). Por ejemplo, podemos obtener el marco de datos 
# agrupados para el año 1967. Comenta tus resultados. Escoge otro año 
# explicando el porqué de la elección y qué resultados obtuviste.

grupo_1967 = gapminder.groupby('year').get_group(1967)
grupo_1967
```

```python
grupo_2007 = gapminder.groupby('year').get_group(2007)
grupo_2007
```

Comentarios:
En la tabla correspondiente al año 2007, se observa un aumento en la población mundial y un aumento en la esperanza de vida, lo que sugiere una mejora en la calidad de vida. Además, se ve un aumento en el ingreso per cápita en muchos países, lo que también sugiere una mejora en la economía mundial en comparación con el año 1967.

```python
# Ahora con los datos anteriores construya tablas pivote según las siguientes 
# instrucciones: 

# 1. Construya una tabla fijando “continent”, con las funciones vistas en clase

# crear tabla 
tabla_1 = gapminder.pivot_table(
    index='continent', # filas
    values=['lifeExp','pop','gdpPercap'],) # columnas

tabla_1
```

```python
# 2. Realice una tabla de doble entrada, fijando “continent” y “year”. Comente sus resultados.

# crear tabla de doble entrada
tabla_2 = gapminder.pivot_table( 
    values='pop', # valores
    index='year', # filas
    columns='continent', # columnas
    aggfunc='mean' # función de agregación
)

tabla_2.head(5)
```

```python
# 3. calcualr el promedio de la variable pop continente

# Agrupar los datos por continente y calcular el promedio de la variable "pop"
promedio_por_contienente = gapminder.groupby('continent')['pop'].mean()
promedio_por_contienente
```

```python
# 4. Obtenga un gráfico de barras del punto 3, comente sus resultados.

# creamos el gráfico de barras
sns.barplot(
    x=promedio_por_contienente.index, 
    y=promedio_por_contienente.values,
    palette='Blues_d') 

# establecemos las etiquetas de eje x y y
plt.xlabel('Continente')
plt.ylabel('Promedio de pop')

# establecemos el título del gráfico
plt.title('Promedio de pop por continente')

# mostramos el gráfico
plt.show()
```

Comentarios:
Podemos ver que el continente con el promedio de población más alto es Asia, seguido de América y Europa. África y Oceanias tiene el promedio de población más bajo.
