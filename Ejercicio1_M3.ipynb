```python
# importar librerias

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# En una clínica infantil se ha ido anotando, durante un mes, el número de metros que cada niño anda, seguido y sin caerse, 
# el primer día que comienza a caminar, obteniéndose la tabla de información adjunta:

# Datos
metros = [1,2,3,4,5,6,7,8]
niños = [2,6,10,5,10,3,2,2]


# a) Tabla de Frecuencia.

# Crear DataFrame con los datos
df = pd.DataFrame({'metros': metros, 'niños': niños})

# Calcular frecuencias absolutas
df['frecuencia absoluta'] = df['niños']

# Calcular frecuencias relativas
df['frecuencia relativa'] = df['niños'] / df['niños'].sum()

# Calcular frecuencias acumuladas
df['frecuencia acumulada'] = df['frecuencia absoluta'].cumsum()

df
```

```python
# graficos de las frecuencias

df.plot.bar(x='metros', y='frecuencia absoluta')
df.plot.bar(x='metros', y='frecuencia relativa')
df.plot.bar(x='metros', y='frecuencia acumulada')
```

```python
# b) Mediana, media aritmética, moda y cuartiles.

# Calcular mediana
mediana = df['metros'].median()
print('Mediana:', mediana)

# Calcular media aritmética
media = df['metros'].mean()
print('Media aritmética:', media)

# Calcular moda
moda = df['metros'].mode()[0]
print('Moda:', moda)

# Calcular cuartiles
cuartiles = df['metros'].quantile([0.25, 0.5, 0.75])
print('Cuartiles:', cuartiles.tolist())
```

```python
# c) Varianza y desviación Estándar.

# Varianza
varianza = df['metros'].var()

# Desviacion Estandar
desviacion_estandar = df['metros'].std()

print("Varianza:", varianza)
print("Desviación Estándar:", desviacion_estandar)
```

```python
# d) ¿Entre qué dos valores se encuentra, como mínimo, el 75% de las observaciones?

df.quantile([0, 0.75])


# El rango que incluye al menos el 75% de las observaciones está entre 1 y 6.5 metros.
# Con respecto a la variable niños esta está entre 2 y 7 niños

```
