```python
# Importar librerías a utilizar para el caso en cuestión (panda, matplotlib)
import matplotlib.pyplot as plt
import pandas as pd
```

```python
## UTILIZAR DATAFRAME ENTREGADO POR PROFESOR ##

#  Leer csv correspondiente.
df = pd.read_csv('HousingData.csv')
df
```

```python
# Renombrar campos.

data = df.rename(columns={
    "CRIM": "indice_crimen", 
    "ZN": "prop_zona_residencial", 
    "INDUS": "prop_industrial",
    "CHAS": "cerca_rio", 
    "NOX": "concentracion_oxido_nitroso", 
    "RM": "num_habitaciones",
    "AGE": "prop_antiguas", 
    "DIS": "dist_centro", 
    "RAD": "acc_autopista", 
    "TAX": "impuesto_propiedad", 
    "PTRATIO": "ratio_profesor", "B": "prop_negra", 
    "LSTAT": "porcentaje_clase_baja", 
    "MEDV": "valor_medio_casas"}, inplace=True)

df
```

```python
# Histograma: Un histograma es útil para ver la distribución de una
# variable, es decir, nos permite ver los valores más comunes. Para
# el caso que se presenta se desea ver la distribución de la cantidad
# media de habitaciones en el estudio realizado.

# Crear un histograma
fig, ax = plt.subplots() # Crear una figura y un conjunto de ejes
ax.hist(df['num_habitaciones'], bins=20) # Crear un histograma con 20 bins

ax.set_xlabel('Cantidad media de habitaciones') # Establecer etiqueta del eje x
ax.set_ylabel('Frecuencia') # Establecer etiqueta del eje y
ax.set_title('Distribución de la cantidad media de habitaciones') # Establecer título

# Mostrar el gráfico
plt.show()
```

```python
# Dispersión: sirve para representar la relación entre dos variables.
# Para nuestro caso se desea ver la relación entre índice de crimen
# y el valor mediano de las casas. En otras palabras, ¿el índice de
# crimen afecta el valor medio de las casas?

plt.scatter(df['indice_crimen'], df['valor_medio_casas']) # Crear un gráfico de dispersión
plt.title('Relación entre índice de crimen y valor mediano de las casas') # Establecer título
plt.xlabel('indice de crimen') # Establecer etiqueta del eje x
plt.ylabel('Valor mediano de las casas') # Establecer etiqueta del eje y
plt.show() # Mostrar el gráfico
```

```python
#Barras: Un gráfico de barras es útil para comparar una variable
# entre distintos grupos o categorías. Por ejemplo, si quisiéramos
# observar el valor medio de cada ciudad. Para nuestro caso se
# desea agrupar por ciudad y toma la media del valor mediano.
# Gráfica las 10 primeras ciudades.

df_ciudades = df.groupby('Ciudad').mean().sort_values(by='valor_medio_casas', ascending=False).head(10)
plt.bar(df_ciudades.index, df_ciudades['valor_medio_casas'])
plt.title('Valor medio de las casas por ciudad (10 primeras)')
plt.xlabel('Ciudad')
plt.ylabel('Valor mediano de las casas')
plt.show()


###############################################################################################################
## NO EXISTE LA COLUMNNA CIUDAD EN ARCHIVO ENTREGADO POR PROFESOR, POR LO QUE NO SE PUEDE EJECUTAR EL CÓDIGO ##
###############################################################################################################
```

```python
# Cajas: Los diagramas de cajas son útiles para representar grupos
# de datos y compararlos entre ellos. Otro uso importante es que
# nos permiten identificar de forma sencilla si una variable tiene
# muchos outliers (valores atípicos) esto es, elementos que se
# alejan de los valores frecuentes de dicha variable. Para nuestro
# caso, se desea ver los valores atípicos de índice de crimen en los
# diferentes cuantiles de valor mediano.

# Crear cuantiles
df['Cuantil'] = pd.qcut(df['valor_medio_casas'], q=4) 

# Crear diagrama de cajas
df.boxplot(column='indice_crimen', by='Cuantil') 

# Establecer título
plt.title('Valores atípicos de índice de crimen en los diferentes cuantiles de valor mediano') 

# Eliminar etiqueta del eje x
plt.suptitle('')

# Mostrar el gráfico
plt.xlabel('Cuantil de valor mediano')
plt.ylabel('Índice de crimen')
plt.show()
```
