```python
#importar librerias

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Una compañía desea hacer predicciones del valor anual de sus ventas totales en cierto país a partir de la relación de éstas 
# y la renta nacional. Para investigar la relación cuenta con los siguientes datos:

# cargar datos en un DataFrame 
df = pd.DataFrame(
    {'renta': [189,190,208,227,239,252,257,254,293,308,316],
     'ventas': [402,404,412,425,429,436,440,447,458,469,469]})

df
```

```python
# 1. La recta de regresión lineal de Y sobre X.

# Ajustar un modelo de regresión lineal
regresion = LinearRegression().fit(df[['renta']], df['ventas'])

# imprimir la pendiente y la ordenada al origen
print('Pendiente:', regresion.coef_[0])
print('Ordenada al origen:', regresion.intercept_)

# graficar datos y recta de regresión lineal
plt.scatter(df['renta'], df['ventas'])
plt.plot(df['renta'], regresion.predict(df[['renta']]), color='red')
plt.xlabel('Renta Nacional (millones de euros)')
plt.ylabel('Ventas (miles de euros)')
plt.title('Regresión lineal de Ventas sobre Renta Nacional')
plt.show()
```

```python
# 2. El coeficiente de correlación lineal e interpretarlo.

# Calcular el coeficiente de correlación de Pearson entre dos columnas
coef_corr = df["renta"].corr(df["ventas"])

# Imprimir el coeficiente de correlación
print("Coeficiente de correlación renta-ventas:", coef_corr)


# INTERPRETACION

# Si aumenta la renta nacional aumenta las ventas, hay una correlación directa.
# 0,99 es una correlación alta, hay una relación muy fuerte y positiva entre las dos variables
```

```python
# 3. Si en 2011 la renta nacional del país fue de 325 millones de euros. 
# ¿Cuál será la predicción para las ventas de la compañía en este año?

# valor de la renta nacional en 2011
renta_2011 = 325

# predicción de ventas en 2011
venta_pred = regresion.predict([[renta_2011]])

print(f"La predicción de ventas en 2011 es de {venta_pred[0]:.2f} miles de euros.")

# El índice [0] se utiliza para acceder al primer elemento del array venta_pred, que es el valor de la predicción de ventas para el año 2011.
```

```python
# INTERPRETACION
# Se estima que la compañía habra vendido aproximadamente 476.829 euros en 2011, dado que la renta nacional en ese año fue de 325 millones de euros. 
```
